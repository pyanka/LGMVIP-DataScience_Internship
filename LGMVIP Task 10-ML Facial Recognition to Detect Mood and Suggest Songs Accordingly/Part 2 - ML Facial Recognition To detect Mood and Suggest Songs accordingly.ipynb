```python
import numpy as np
import pandas as pd
import matplotlib .pyplot as plt
import os
import warnings
warnings.filterwarnings('ignore')
from mtcnn import MTCNN
from tensorflow.keras.models import load_model
```

```python
model = load_model('C:/Users/15DBCOM40/Desktop/Internship Projects/LGMVIP Task 10-ML Facial Recognition to Detect Mood and Suggest Songs Accordingly/model_optimal.h5')
print('model Loaded 100%')
```

```python
img = 'https://static-bebeautiful-in.unileverservices.com/Unlock-flawless-skin_MobileHomeFeature.jpg'
try: 
    image = plt.imread(img)
except SyntaxError:
    image = plt.imread(img, format='jpg')
plt.imshow(image)    
    
```

```python
detect = MTCNN()
results = detect.detect_faces(image)

x1,y1,width,height = results[0]['box']
x2,y2 = x1+width,y1+height

face = image[y1:y2,x1:x2]
plt.imshow(face)
```

```python
test_image = np.resize(face,(224,224,1))
```

```python
label_dict = {0:'Angry',1:'Disgust',2:'Fear',3:'Happy',4:'Natural',5:'Sad',6:'Surprise'}
test_img = np.expand_dims(test_image,axis=0)
test_img = test_img.reshape(1,224,224,1)
result = model.predict(test_img)
result = list(result[0])
img_index = result.index(max(result))
emotion = label_dict[img_index]
print(emotion)
```
